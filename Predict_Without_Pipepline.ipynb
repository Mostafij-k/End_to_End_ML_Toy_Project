```python
# Import required libraries
import numpy as np
import pickle
```

```python
# Import ML model object
ohe_sex=pickle.load(open('models/ohe_sex.pkl', 'rb'))
ohe_embarked=pickle.load(open('models/ohe_embarked.pkl', 'rb'))
clf=pickle.load(open('models/clf.pkl', 'rb'))
```

```python
# Take user input
#Pclass/gender/age/sibSp/parch/fare/embarked
user_input=np.array([2,'male', 31.0, 0, 0, 10.5, 'S'], dtype=object).reshape(1,7)
```

```python
# Show user_input
user_input
```

```python
# Now perform OneHotEncoding on Sex and Ebarked column
user_input_sex=ohe_sex.transform(user_input[:,1].reshape(1,1))
user_input_embarked=ohe_embarked.transform(user_input[:,-1].reshape(1,1))
```

```python
# Extract user input age column
user_input_age=user_input[:, 2].reshape(1,1)
```

```python
# Make final transformed column
user_input_transformed=np.concatenate(
    (user_input[:,[0,3,4,5]], 
    user_input_age,
    user_input_sex,
    user_input_embarked),
    axis=1
)
```

```python
# Shape of user_input_transformed 
user_input_transformed.shape
```

```python
# Now make prediction
clf.predict(user_input_transformed)

# Passenger survived
```
